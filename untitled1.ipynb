```python
cases = data_russia.filter(['new_cases'])
# сделаем array
cases = cases.values
#Get the number of rows to train the model on
training_data_len = math.ceil( len(cases) * 0.8 )

training_data_len
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_cases = scaler.fit_transform(cases)
plt.plot(scaled_cases)
```

```python
#Create the training data set 
#Create the scaled training data set
cases_train = scaled_cases[0:training_data_len, :]
#Split the data into x_train and y_train data sets
x_train = []
y_train = []
#We create a loop
for i in range(60, len(cases_train)):
  x_train.append(cases_train[i-60:i, 0]) #Will conaint 60 values (0-59)
  y_train.append(cases_train[i, 0]) #Will contain the 61th value (60)
  '''
  if i <= 60:
    print(x_train)
    print(y_train)
    print()
  '''

#Convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)


#Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

#Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

#Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)

#Create the testing data set
#Create a new array containing scaled values from index 1738 to 2247
test_data = scaled_cases[training_data_len - 60:]
#Create the data set x_test and y_test
x_test = []
y_test = cases[training_data_len:, :]
for i in range(60, len(test_data)):
  x_test.append(test_data[i-60:i, 0])

#Convert the data to a numpy array
x_test = np.array(x_test)

#Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

#Get the model's predicted price values for the x_test data set
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
predictions

#Evaluate model (get the root mean quared error (RMSE))
rmse = np.sqrt( np.mean( predictions - y_test )**2 )
rmse

#Plot the data
train = data_russia[:training_data_len]
valid = data_russia[training_data_len:]
valid['Predictions'] = predictions
#Visualize the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['new_cases'])
plt.plot(valid[['new_cases', 'Predictions']])
plt.legend(['Train', 'Validation', 'Predictions'], loc='lower right')
plt.show()
```
